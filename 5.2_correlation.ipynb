```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'dataset_allmerged_0801.csv'
data = pd.read_csv(file_path)

data.head()
```

```python
data.describe()
```

```python
ndvi_columns = [col for col in data.columns if 'mean_ndvi' in col]
```

```python
# List all climate variables in the dataset
all_climate_variables = [
    col for col in data.columns if col not in ndvi_columns and col not in ['date', 'Unnamed: 0']
]

# Define a function to compute and format the correlation for a given NDVI column
def compute_correlation(ndvi_column):
    correlation_with_ndvi = data[all_climate_variables].corrwith(data[ndvi_column]).sort_values(ascending=True)
    correlation_df = correlation_with_ndvi.reset_index()
    correlation_df.columns = ['Climate Variable', f'Correlation with {ndvi_column}']
    clean_correlation_df = correlation_df.dropna()
    return clean_correlation_df

# Compute the correlation for each NDVI region
correlation_ndvi_BD = compute_correlation('mean_ndvi_BD')
correlation_ndvi_FN = compute_correlation('mean_ndvi_FN')
correlation_ndvi_FS = compute_correlation('mean_ndvi_FS')
correlation_ndvi_SF = compute_correlation('mean_ndvi_SF')

correlation_ndvi_SF.head(27) 
```

```python
top_27_ndvi_BD = set(correlation_ndvi_BD.head(27)['Climate Variable'])
top_27_ndvi_FN = set(correlation_ndvi_FN.head(27)['Climate Variable'])
top_27_ndvi_FS = set(correlation_ndvi_FS.head(27)['Climate Variable'])
top_27_ndvi_SF = set(correlation_ndvi_SF.head(27)['Climate Variable'])

# intersection of these sets
common_climate_variables = top_27_ndvi_BD & top_27_ndvi_FN & top_27_ndvi_FS & top_27_ndvi_SF
common_climate_variables
```

```python
# NEgative

def filter_common_variables(correlation_df, common_variables):
    return correlation_df[correlation_df['Climate Variable'].isin(common_variables)]

filtered_correlation_ndvi_BD = filter_common_variables(correlation_ndvi_BD, common_climate_variables)
filtered_correlation_ndvi_FN = filter_common_variables(correlation_ndvi_FN, common_climate_variables)
filtered_correlation_ndvi_FS = filter_common_variables(correlation_ndvi_FS, common_climate_variables)
filtered_correlation_ndvi_SF = filter_common_variables(correlation_ndvi_SF, common_climate_variables)

filtered_correlation_ndvi_BD.set_index('Climate Variable', inplace=True)
filtered_correlation_ndvi_FN.set_index('Climate Variable', inplace=True)
filtered_correlation_ndvi_FS.set_index('Climate Variable', inplace=True)
filtered_correlation_ndvi_SF.set_index('Climate Variable', inplace=True)

combined_correlation = pd.concat([
    filtered_correlation_ndvi_BD,
    filtered_correlation_ndvi_FN,
    filtered_correlation_ndvi_FS,
    filtered_correlation_ndvi_SF
], axis=1)
combined_correlation.columns = ['mean_ndvi_BD', 'mean_ndvi_FN', 'mean_ndvi_FS', 'mean_ndvi_SF']

# Plot
plt.figure(figsize=(12, 10))
sns.heatmap(combined_correlation, annot=True, cmap='BrBG')
plt.title('Correlation Heatmap of Common Climate Variables with NDVI for All Regions')
plt.show()
```
